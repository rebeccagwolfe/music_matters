```python
import requests
import json
import numpy as np
import pandas as pd
import time

with open('../Data/api_key.json') as fi:
    credentials = json.load(fi)
    
api_key = credentials['api_key']

header={
    "Accept":"application/json", 
    "x-api-key":api_key
}


beatles=requests.get("https://api.setlist.fm/rest/1.0/artist/b10bbbfc-cf9e-42e0-be17-e2c3e1d2600d/setlists?p=1", headers=header)
print(beatles)
```

```python
beatles_test = beatles.json()
beatles_test
```

```python
beatles_test['setlist']
```

```python
beatles_test['setlist'][3]['venue']['id']
```

```python
beatles_test['setlist'][0]['sets']['set'][0]['song'][0]['name']
```

```python

ids = []
versionid = []
eventdate = []
lastupdated = []
artist = []
venue = []
venue_id = []
song = []
cover = []
songinfo = []
for x in beatles_test['setlist']:
    for y in x['sets']['set'][0]['song']:
        song.append(y['name'])
        ids.append(x['id'])
        versionid.append(x['versionId'])
        eventdate.append(x['eventDate'])
        lastupdated.append(x['lastUpdated'])
        artist.append(x['artist']['name'])
        venue.append(x['venue']['name'])
        venue_id.append(x['venue']['id'])
        try: 
            cover.append(y['cover']['name'])
        except: 
            cover.append("original")
beatles_setlist = pd.DataFrame({'id': ids, 'versionid': versionid, 'eventdate': eventdate, 'lastupdated': lastupdated, 'artist': artist, 'venue': venue, 'venue_id':venue_id, 'song': song, 'cover': cover})
beatles_setlist.head(50)
```
