## Code to use API call to get venues per state with location information- then seperate API call to get arists ids. Will then loop through artist ids in another API call to get set list information/songs based by looping through atist ids (mbid)

```python
import requests
import json
import numpy as np
import pandas as pd
import time

with open('../Data/api_key.json') as fi:
    credentials = json.load(fi)
    
api_key = credentials['api_key']

header={
    "Accept":"application/json", 
    "x-api-key":api_key
}
```

```python
# This loop is running all pages of venues for the first five states
colorado_venues= []

states=['CO']

# if print request peice = response to 100, print venues.append...else break
# need to figure out how many pages of venues each state has. may need try and accept. or while loop.

for y in states: 
    z=1
    a=1
    for x in range(10000):
        if str(requests.get(f"https://api.setlist.fm/rest/1.0/search/venues?p={x+1}&stateCode={y}", headers=header))=='<Response [200]>':
            print(requests.get(f"https://api.setlist.fm/rest/1.0/search/venues?p={x+1}&stateCode={y}", headers=header))
            colorado_venues.append(requests.get(f"https://api.setlist.fm/rest/1.0/search/venues?p={x+1}&stateCode={y}", headers=header).json())
            time.sleep(5)
            print(f"try:{z}")
            z+=1
        else:
            print(f"break:{a}")
            a+=1
            break
                   
```

```python
# if I change the [0] to 1, it shows me the next page. I can keep doing this until 16 to show me page 17. 
# Is there a way to make a loop? Thinking I can assign it to y and say "for y in range (1,16):" 

venue_id = []
names = []
cities = []
city_id = []
states = []
lat_coords = []
long_coords =[]
country = []
for y in range(len(colorado_venues)):    
    for x in colorado_venues[y]['venue']:
        venue_id.append(x['id'])
        names.append(x['name'])
        cities.append(x['city']['name'])
        city_id.append(x['city']['id'])
        states.append(x['city']['state'])
        lat_coords.append(x['city']['coords']['lat'])
        long_coords.append(x['city']['coords']['long'])
        country.append(x['city']['country']['code'])
colorado_venues_df = pd.DataFrame({'venue_id':venue_id, 'name':names, 'city':cities,'city_id':city_id, 'state':states, 'lat':lat_coords, 'long':long_coords, 'country':country})
colorado_venues_df.head(10)
```

```python
colorado_venue_id_list

colorado_artists_list = []

for x in colorado_venue_id_list:
    print(requests.get(f"https://api.setlist.fm/rest/1.0/venue/{x}/setlists?p=1", headers=header))
    colorado_artists_list.append(requests.get(f"https://api.setlist.fm/rest/1.0/venue/{x}/setlists?p=1", headers=header).json())
    time.sleep(1)
```

```python
mbid = []
setlist_id = []
artist_name = []
venue_name = []
venue_id = []

for y in range(len(colorado_artists_list)):
    try:
        for x in artists_list[y]['setlist']:
            mbid.append(x['artist']['mbid'])
            artist_name.append(x['artist']['name'])
            setlist_id.append(x['id'])
            venue_name.append(x['venue']['name'])
            venue_id.append(x['venue']['id'])
    except:
        for x in artists_list[y]:
            mbid.append("not found")
            artist_name.append("not found")
            setlist_id.append("not found")
            venue_name.append("not found")
            venue_id.append("not found")
colorado_artists_df = pd.DataFrame({'mbid':mbid, 'setlist_id':setlist_id, 'artist_name':artist_name, 'venue_name':venue_name, 'venue_id':venue_id})
colorado_artists_df.head(10)
```
