```python
# made a dataframe using data pulled from setlist by venue, but don't know how to get the information for each song into a df. 

ids = []
versionid = []
eventdate = []
lastupdated= []
artist = []
venue = []
venue_id = []
song = []
cover = []

for x in red_rocks_venue['setlist']:
    ids.append(x['id'])
    versionid.append(x['versionId'])
    eventdate.append(x['eventDate'])
    lastupdated.append(x['lastUpdated'])
    artist.append(x['artist']['name'])
    venue.append(x['venue']['name'])
    venue_id.append(x['venue']['id'])
    try:
        song.append(x['sets']['set'][0]['song'][0]['name'])
        try:
            cover.append(x['sets']['set'][0]['song'][0]['cover']['name'])
        except:
            cover.append("original")
    except:
        song.append("not_found")  
        cover.append('not_found')
red_rocks_setlist = pd.DataFrame({'ids':ids, 'versionid':versionid, 'eventdate':eventdate, 'lastupdated':lastupdated, 'artist':artist, 'venue':venue, 'venue_id':venue_id, 'song':song, 'cover':cover})
red_rocks_setlist.head(50)
```
