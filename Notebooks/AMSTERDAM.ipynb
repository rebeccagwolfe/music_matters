```python
with open('../Data/api_key.json') as fi:
    credentials = json.load(fi)
    
api_key = credentials['api_key']

header={
    "Accept":"application/json", 
    "x-api-key":api_key
}
```

```python
amsterdam= []


city = 'Amsterdam'
countryCode = 'NL'

# https://api.setlist.fm/rest/1.0/search/setlists?cityName=brussels&countryCode=FR&p=1&year=2022 

# need to specify in artists api request the city, state, and year

 
z=1
a=1
for x in range(113):
    url = requests.get(f"https://api.setlist.fm/rest/1.0/search/setlists?cityName={city}&countryCode={countryCode}&p={x+1}&year=2022", headers=header)
    if str(url)=='<Response [200]>':
        print(url) 
        amsterdam.append(url.json())
        time.sleep(5)
        print(f"try:{z}")
        z+=1
    else:
        print(f"break:{a}")
        a+=1
        break
```

```python
# skip right to dataframe

ids = []
versionid = []
eventdate = []
lastupdated = []
artist = []
venue = []
venue_id = []
song = []
cover = []
cities = []
city_id = []
states = []
countries = []
countryCode =[]
stateCode = []
lat_coords = []
long_coords =[]
country = []
z=0
for a in amsterdam:
    try:
        for x in a['setlist']:
            try:
                for y in x['sets']['set'][0]['song']:
                    song.append(y['name'])
                    ids.append(x['id'])
                    versionid.append(x['versionId'])
                    eventdate.append(x['eventDate'])
                    lastupdated.append(x['lastUpdated'])
                    artist.append(x['artist']['name'])
                    try:
                        venue.append(x['venue']['name'])
                        venue_id.append(x['venue']['id'])
                    except:
                        venue.append("not_found")
                        venue_id.append("not_found")
                    cities.append(x['venue']['city']['name'])
                    states.append(x['venue']['city']['state'])
                    countries.append(x['venue']['city']['country']['name'])
                    stateCode.append(x['venue']['city']['stateCode'])
                    countryCode.append(x['venue']['city']['country']['code'])
                    lat_coords.append(x['venue']['city']['coords']['lat'])
                    long_coords.append(x['venue']['city']['coords']['long'])
                    try: 
                        cover.append(y['cover']['name'])
                    except: 
                        cover.append("original")
                z+=1
            except:
                z+=1
                continue
    except:
        continue
amsterdam_setlists_df = pd.DataFrame({'ids': ids, 'versionid': versionid, 'eventdate': eventdate, 'lastupdated': lastupdated, 'artist': artist, 'venue': venue, 'venue_id':venue_id,'city':cities,'state':states, 'country':countries, 'stateCode':stateCode, 'countryCode':countryCode,'lat_coords':lat_coords, 'long_coords':long_coords, 'song': song, 'cover': cover})
amsterdam_setlists_df.head(50)
```
