```python
import requests
import json
import numpy as np
import pandas as pd
import time

with open('../Data/api_key.json') as fi:
    credentials = json.load(fi)
    
api_key = credentials['api_key']

header={
    "Accept":"application/json", 
    "x-api-key":api_key
}
```

```python
# venues list of only 5 states AND replaced .text with .json() at the end so the list is already converted to jsons.

header={
    "Accept":"application/json", 
    "x-api-key":api_key
}

state={
    "stateCode":"TN"
}


states=['AL','AK','AZ','AR','CA']

venues_list = []

for x in states:
    print(requests.get(f"https://api.setlist.fm/rest/1.0/search/venues?p=1&stateCode={x}", headers=header))
    venues_list.append(requests.get(f"https://api.setlist.fm/rest/1.0/search/venues?p=1&stateCode={x}", headers=header).json())
    time.sleep(10)
```

```python
# found out how to the venues in dataframe, but only one state at a time. These are Alabama

ids = []
names = []
cities = []
city_id = []
states = []
lat_coords = []
long_coords =[]
country = []
for x in venues_list[0]['venue']:
    ids.append(x['id'])
    names.append(x['name'])
    cities.append(x['city']['name'])
    city_id.append(x['city']['id'])
    states.append(x['city']['state'])
    lat_coords.append(x['city']['coords']['lat'])
    long_coords.append(x['city']['coords']['long'])
    country.append(x['city']['country']['code'])
alabama_venues = pd.DataFrame({'id':ids, 'name':names, 'city':cities,'city_id':city_id, 'state':states, 'lat':lat_coords, 'long':long_coords, 'country':country})
alabama_venues.head(10)
```
