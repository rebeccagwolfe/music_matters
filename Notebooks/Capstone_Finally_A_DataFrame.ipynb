```python
import requests
import json
import numpy as np
import pandas as pd
import time

with open('../Data/api_key.json') as fi:
    credentials = json.load(fi)
    
api_key = credentials['api_key']
```

```python
header={
    "Accept":"application/json", 
    "x-api-key":api_key
}


b=requests.get("https://api.setlist.fm/rest/1.0/artist/b10bbbfc-cf9e-42e0-be17-e2c3e1d2600d/setlists?p=1", headers=header)
print(b)
```

```python
b_test = b.json()
ids = []
versionid = []
eventdate = []
lastupdated = []
artist = []
venue = []
song = []
for x in b_test['setlist']:
    for y in x['sets']['set'][0]['song']:
        song.append(y['name'])
        ids.append(x['id'])
        versionid.append(x['versionId'])
        eventdate.append(x['eventDate'])
        lastupdated.append(x['lastUpdated'])
        artist.append(x['artist']['name'])
        venue.append(x['venue']['name'])
b_setlist_test = pd.DataFrame({'id': ids, 'versionid': versionid, 'eventdate': eventdate, 'lastupdated': lastupdated, 'artist': artist, 'venue': venue, 'song': song})
b_setlist_test
```
