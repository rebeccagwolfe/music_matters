## API REQUEST 1


this request gives clean df with all songs in setlist

```python
austin= []


city = 'Austin'
stateCode = 'TX'
   
# f"https://api.setlist.fm/rest/1.0/search/setlists?cityName={city}&p={x+1}&stateCode={stateCode}&year=2022"    

# need to specify in artists api request the city, state, and year

for y in stateCode: 
    z=1
    a=1
    for x in range(25):
        if str(requests.get(f"https://api.setlist.fm/rest/1.0/search/setlists?cityName={city}&p={x+1}&stateCode={stateCode}&year=2022", headers=header))=='<Response [200]>':
            print(requests.get(f"https://api.setlist.fm/rest/1.0/search/setlists?cityName={city}&p={x+1}&stateCode={stateCode}&year=2022", headers=header))
            austin.append(requests.get(f"https://api.setlist.fm/rest/1.0/search/setlists?cityName={city}&p={x+1}&stateCode={stateCode}&year=2022", headers=header).json())
            time.sleep(5)
            print(f"try:{z}")
            z+=1
        else:
            print(f"break:{a}")
            a+=1
            break
                   
```

```python
# skip right to dataframe

ids = []
versionid = []
eventdate = []
lastupdated = []
artist = []
venue = []
venue_id = []
song = []
cover = []
cities = []
city_id = []
states = []
stateCode = []
lat_coords = []
long_coords =[]
country = []
z=0
for a in austin:
    try:
        for x in a['setlist']:
            try:
                for y in x['sets']['set'][0]['song']:
                    song.append(y['name'])
                    ids.append(x['id'])
                    versionid.append(x['versionId'])
                    eventdate.append(x['eventDate'])
                    lastupdated.append(x['lastUpdated'])
                    artist.append(x['artist']['name'])
                    venue.append(x['venue']['name'])
                    venue_id.append(x['venue']['id'])
                    cities.append(x['venue']['city']['name'])
                    states.append(x['venue']['city']['state'])
                    stateCode.append(x['venue']['city']['stateCode'])
                    lat_coords.append(x['venue']['city']['coords']['lat'])
                    long_coords.append(x['venue']['city']['coords']['long'])
                    try: 
                        cover.append(y['cover']['name'])
                    except: 
                        cover.append("original")
                z+=1
            except:
                z+=1
                continue
    except:
        continue
austin_setlists_df_2 = pd.DataFrame({'ids': ids, 'versionid': versionid, 'eventdate': eventdate, 'lastupdated': lastupdated, 'artist': artist, 'venue': venue, 'venue_id':venue_id,'city':cities,'state':states, 'stateCode':stateCode, 'lat_coords':lat_coords, 'long_coords':long_coords, 'song': song, 'cover': cover})
austin_setlists_df_2.head(50)
```
