# Obtaining artist id (mbid) from set list call that requires venue_id. Since I have a dataframe of venue_ids form all states. 

```python
import requests
import json
import numpy as np
import pandas as pd
import time

with open('../Data/api_key.json') as fi:
    credentials = json.load(fi)
    
api_key = credentials['api_key']

header={
    "Accept":"application/json", 
    "x-api-key":api_key
}
```

```python
venue_id = '3d259ab'

    
the_end_venue = requests.get(f"https://api.setlist.fm/rest/1.0/venue/{venue_id}/setlists?p=1", headers=header)
print(the_end_venue)
```

```python
the_end_venue_json = the_end_venue.json()
the_end_venue_json
```

```python
the_end_venue_df = pd.DataFrame(the_end_venue_json)
the_end_venue_df
```

```python
the_end_venue_df['setlist'][0]['venue']
```

```python
the_end_venue_data = the_end_venue_df['setlist'].apply(pd.Series)
the_end_venue_data.head()
```

```python
# the_end_venue_df['setlist'][0]['artist']['mbid']

mbid = []
setlist_id = []
artist_name = []
venue_name = []
venue_id = []

for x in the_end_venue_df['setlist']:
    mbid.append(x['artist']['mbid'])
    artist_name.append(x['artist']['name'])
    setlist_id.append(x['id'])
    venue_name.append(x['venue']['name'])
    venue_id.append(x['venue']['id'])
the_end_venue_artists = pd.DataFrame({'mbid':mbid, 'setlist_id':setlist_id, 'artist_name':artist_name, 'venue_name':venue_name, 'venue_id':venue_id})
the_end_venue_artists.head(10)
```
